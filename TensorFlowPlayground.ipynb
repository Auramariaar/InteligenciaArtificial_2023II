# Ejemplo 1: Clasificación de formas

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Crear datos de círculos y cruces
np.random.seed(0)
points = np.random.rand(100, 2)
labels = (np.sum(points**2, axis=1) < 0.7**2).astype(int)

# Crear el modelo
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(2,)),
    tf.keras.layers.Dense(4, activation='tanh'),
    tf.keras.layers.Dense(2, activation='tanh'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Compilar el modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Entrenar el modelo
history = model.fit(points, labels, epochs=100, verbose=0)

# Visualizar el resultado
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.scatter(points[:, 0], points[:, 1], c=labels, cmap='coolwarm')
plt.title("Datos de Entrada")
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.title("Pérdida durante el Entrenamiento")
plt.show()
```

# Ejemplo 2: Regresión de Función Cuadrática

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Crear datos de función cuadrática
np.random.seed(0)
x = np.random.rand(100, 1) * 2 - 1
y = x**2 + np.random.randn(100, 1) * 0.1

# Crear el modelo
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Compilar el modelo
model.compile(optimizer='adam', loss='mean_squared_error')

# Entrenar el modelo
history = model.fit(x, y, epochs=200, verbose=0)

# Visualizar el resultado
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.scatter(x, y)
plt.title("Datos de Entrada")
plt.xlabel("X")
plt.ylabel("Y")
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.title("Pérdida durante el Entrenamiento")
plt.xlabel("Época")
plt.ylabel("Pérdida")
plt.show()
```

# Ejemplo 3: Detección de Ruido en Datos

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Crear datos con ruido gaussiano
np.random.seed(0)
x = np.linspace(0, 1, 100)
y = np.sin(2 * np.pi * x) + np.random.normal(0, 0.3, 100)

# Crear el modelo
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(2, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Compilar el modelo
model.compile(optimizer='adam', loss='mean_squared_error')

# Entrenar el modelo
history = model.fit(x, y, epochs=200, verbose=0)

# Visualizar el resultado
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.scatter(x, y)
plt.plot(x, model.predict(x), color='red', linewidth=3)
plt.title("Datos de Entrada con Ruido")
plt.xlabel("X")
plt.ylabel("Y")
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.title("Pérdida durante el Entrenamiento")
plt.xlabel("Época")
plt.ylabel("Pérdida")
plt.show()
```
